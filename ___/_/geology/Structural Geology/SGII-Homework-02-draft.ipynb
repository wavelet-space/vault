The rock containing trilobite fossil is affected by deformation defined by deformation gradient

$$\mathbf{F}=\begin{bmatrix}2 & 1\\ 0 & 0.5\end{bmatrix}$$

The orientation of the undeformed trilobite is defined by angle $\theta$ measured clockwise from x-axis of our reference frame as show on figure.


1. For what angle(s) $\theta$ the deformed trilobite will preserve his bilateral symmetry?
2. What will be the axial ratio of deformed trilobite for calculated angle(s) $\theta$?<br/>
   Assume that the axial ratio of undeformed trilobite is **1.5**

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.radians(30)

# Deformation gradient tensor
F = np.array([[2, 1], [0, 0.5]])

# Rotation tensor
R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

e1 = np.array([1, 0])
e2 = np.array([0, 1])
```
