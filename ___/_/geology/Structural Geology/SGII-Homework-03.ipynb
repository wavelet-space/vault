## From velocity to strain-rate

Imagine zone of deformation with width $d=100$km, with left side fixed and right side moving with velocity $v=1$cm/year. What is the strain-rate $\dot{\varepsilon}_x$ in horizontal direction?
You already know that strain-rate could be calculated from stretch and time:

$$\dot{\varepsilon}_x = \frac{\ln(S_x)}{\Delta t}$$

Use this equation to calculate strain-rate for various times of deformation ranging from 10000 years to 1 year. With decreasing time, your calculations should converge to actual strain-rate.

1. Make plot of time vs. strain-rate (use logarithmic scaling for time)
2. From what you calculated, is it correct that $\dot{\varepsilon}_x = \frac{v}{d}$?

*The reason that strain-rate vary for different times is that during constant strain-rate deformation the velocity is not constant.*
