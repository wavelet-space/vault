## 1. Create deformation gradient
Define isochoric deformation gradient `F` which is characterized by simple shear $\gamma=2$ along line inclined 35° anticlockwise from x-axis. It should represents deformation along ADD shear zone on the image. NOTE THE SENSE OF SHEAR!!!

<img src="https://geosociety.files.wordpress.com/2013/08/jessup_gsa_fig_7.jpg" alt="ADD sheat zone" style="width: 480px;"/>

```python
theta = radians(35)
```

```python
# Simple shear implies isochoric deformation gradient. 
γ = 2.0 

R = array([
    [cos(theta), -sin(theta)],
    [sin(theta), cos(theta)],
    ])

V = array([
    [1, γ], 
    [0, 1],
])

F = V @ R
```

### Check if it is correct

This code will check, whether your matrix `F` is correct. Should show normal shearing along inclined line.

```python
xc, yc = [-3, -3, 3, 3, -3], [-1, 1, 1, -1, -1]
xc, yc = dot(R, [xc, yc])
x, y = dot(F, [xc, yc])

X, Y = meshgrid(
    linspace(-4, 4, 21), 
    linspace(-3, 3, 17)
)

u, v = tensordot(F - eye(2), [X, Y], axes=1)

quiver(X, Y, u, v, angles='xy')

plot(xc, yc, 'r', x, y, 'g')

title('My approximation of ADD shear zone')

axis('equal');
```
