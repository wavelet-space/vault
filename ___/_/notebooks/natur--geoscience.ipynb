## Gravitační pole

__Gravitační zákon__

*Newtonův gravitační zákon* říká, že dvě tělesa se vzájemně přitahují silou podle rovnice   

$$
\vec{F_{12}} = G \frac{m_1 m_2}{r^3}\vec{r_{12}} = G \frac{m_1 m_2}{r^3}r\hat{r}_{12} = G \frac{m_1 m_2}{r^2}\hat{r}_{12}
$$

Tělesa se přitahují vzájemně, tedy velikost sil, jejich směr je ale opačný $\vec{F_{12}}= -\vec{F_{21}}$


Pro velikosti sil obou sil platí, že $F_{12} = F_{21} = F$ a můžeme tedz jednoduše psát

$$
F = F_{12} = F_{2}= G \frac{m_1 m_2}{r^2}.
$$

Rovnice platí pro hmotné body, protože vzdálenost mezi nimi konstantou.

Pro tělesa, která se dají hmotnými body nahradit např. sice nehomogenní tělesa ale při velké vzdálenosti.
Nebo nehomogenní tělesa, kde se však hustota mění radiálně -- ve "slupkách". Jinak bychom museli počítat příspěvek pro proměnlivou vzdálenost pomocí integrálů a také počítáme z něproměnlivou hustotou a tedy i hmotností.

__3. Gravitační zrychlení__ $\vec{g}$

Pokud se zaměříme na jedno z těles např. to s hmotností $m_1$, pak podle zákona síly je $\vec{F} = m\vec{a}$ a tedy z gravitačního zákona získáme úpravami 

$$
G\frac{m_1m_2}{r^2} = m_1\vec{a_1}
$$

, ze které, po vydělení obos stran výrazem $m_1$ vidíme, že zrychlení $\vec{a}$ je uměrné podílu gravitační konstanty a čtverci vzdálenosti mezi hmotnými body. Tomuto zrychlení říkáme gravitační a značíme jako $\vec{g}$ směřuje do středu Země.

$$
\vec{g} = G\frac{m_2}{r^2}
$$

Abychom necvičili s indexy můžeme pro případ měřící sondy o hmotnosti $m$ a Země o hmotnost $M$ zapsat jednoduše:

$$
F_g = mg = G\frac{mM}{r^2}\therefore g = G\frac{M}{r^2}   
$$    

__Proč gravitační síla není vhodná pro popis gravitačního pole a kterou fyzikální veličinu pro popis gravitačního pole Země používáme?__

Důležité je si uvědomit, že s výrazu vypadne hmotnost samostného tělesa (sondy)!
Pro popis gravitačního pole používáme gravitační zrychlení, namísto síly, protože ve výrazu nevystupuje váha sondy.

__Princip superpozice sil__

Princip superpozice sil říká, že výsledný vektor síly působící na hmotný bod je roven součtu všech vektorů sil, které na hmotný bod působí. Pokud je výslednice sil nulová, pak je podle prvního Newtnova zákona hmotný bod v klidu nebo rovnoměrném přímočarém pohybu.  

$$
\vec{F} = \vec{F_1} + \vec{F_2} + \sum\vec{F_i}

__1. Popište konstituční vztah pro napětí a deformaci.__

- Hookův zákon
- mez pružnosti
- pružná deformace

- permanentní deformace

*Konstituční vztah* vzjadřuje závislost deformace vzorku na napětí, kterému je vzorek vystaven.

Každý materiál má svá specifika a proto takový graf vypadá jinak pro různé materiály. Nicméně co mají materiály společného je první lineární část, kterou popisuje tzv. *Hookův zákon* viz dále. Tam kde končí tato lineární závislost, tam se nachází *mez uměrnosti*. Pokud začneme vzorek zatěžovat ještě více pak již závislost deformace na napětí není lineární, ale stále se pohybujeme v oblasti tzv. elastické (pružné) deformace -- vzorek se navrací do původního tvaru. Za touto mazí, která se nazývá *mezí deformace* nastává tzv. plastická (nepružná) deformace --  vzorek se nenavrací do původního tvaru nebo jen částečně. 

Hookův zákon se dá pro malé deformace a malá napětí zapsat pro:

__Tah/Tlak__

$$
\epsilon = \frac{\Delta L}{L} = \frac{\sigma}{E}
$$
, kde $\epsilon$ je poměrné prodloužení, $E$ je *modul pružnosti v tahu/tlaku* (Youngův) $[GPa]$, $\Delta L = L - L_0$.

__Smyk__

$$
\gamma = \frac{\tau}{G}
$$
, kde $\gamma$ je úhel smyku a $G$ je *modul pružnosti ve smyku*.

__2. Jak popisujeme napětí a deformaci na zlomu__

Zlom je ploha oddělující dva bloky horniny.

Napětí je kolmá složka síly působící na danou plochu a obecně se dá vyjářit pomocí skalárního součinu  

$$T = \vec{F}\cdot \vec{S} = FS\cos(\phi),$$

kde $F$ je síla a $S$ je vektor kolmý na plochu o velikosti obshu této plochy. V případě, že je síla kolmá na plochu (vektoru jsou paralelní), přejde vzorez jednoduše na tvar 

$$T = FS\cos(0) = FS$$

Pokud síla působí proti sobě a materiál se tedy ve směru působění stlačuje se tomuto napětí říká *tlak*, v opačném případě o *tah*.

__Napětí__ v nějakém bodě materiálu (horniny) se dá popsat na malé krychliče o rozměrech $dx, dy, dz$. Pro každou plochu této krychličky potřebujeme zapsat zapsat napětí které na ní působí, např. jestli síla působí na plochu z vnitřní strany ven nebo naopak dovnitř -- tím popisujem tah/tlak. Ke každému takovému vektoru potřebujeme pro každou plochu ještě dva na sebe kolmé vektory představující smykové tření. Pro každou plochy tedy po

__5. Fermatův a Hughensuv princip__

__Fermatův princip__ -- Paprsek se šíří prostředím prostředím tak aby dráhu mezi body urazil za co nejmenší čas. V případě izotropního prostředí je to přímka.  Lze z něj odvodit Snellův zákon o odrazu a lomu.

__Huygensův princip__ -- V každém okamžiku lze každý bod na čele vlny chápat jako nový zdroj vlny. Nový tvar vlny pak vytváří obálku. Záíkon není zcela přesný a doplnil ho Fresnel viz Hughensuův Fresnelův princip.

__Princip superpozice__ -- Vlny se šíří prostředím nezávisle na sobě


__6/7/9. Popiště druhy a vlastnosti seismických vln__ 

- **P-Wave** -- Vlnění probíhá ve směru šíření, tzv. *primární* či *podélné* vlny.
  Jde o shuštění a zředění ve směru šíření podobně jako zvuková vlna. Prochází celým zemským tělesem. Jsou to nejrychléjší vlny, detekujemen je tedy jako první. Nezplsobují velké škody, jsou podobné rázové zvukové vlně.  

- **S-Wave** -- Vlnění probíhá ve směru kolmém na směr šíření, tzv. *sekundární* či *příčné* vlny.

__5. Popis sesmických vln pomocí paprsků a vlnoploch__

Seismická vlna se šíří od ohniska všemi směry

- *Paprsek* je pomyslná spojnice od zdroje do místa záznamu zemětřesení. V homogenním prostředí jsou paprsky přímé (nedochází k lomu na rozhranních s různou hustotou). Reálně se ale paprsky zakřivují, protože prochází prostředím s měnící se hustotou.

- *Vlnoplocha* má v homogenním prostředí tvar kulové plochy (přo vělkém poloměru, ve velklé vzdálenost aproxijume již plochou). Vlnoplochy a paprsky jsou v homogenním prostředí na sebe kolmé.

__6. Lom vln na rozhraní -- Snellův zákon__

Seismické vlny se lámou při průchodu na rozhraní prostředí o různých hustotách, která se obecně mění s hloubkou. Pokud vlna přichází z prostředí o menší hustotě do prostředí s větší hustotou, láme se od pomyslné kolmice k rozhraní prostředí a naopak pokud prichází z hustšího do méně hustého protředí, láme se ke kolmici. Proto nejsou paprsky šíření obvykle přímé, ale jsou to křivky viz náčrt.

Seismická vlna ale též nemusí procházet, ale může se odrazit, pokud je překonán tzv. kritický úhel. Analogie je podobná jako s odrazem světla od vodní hladiny.

Nedostatech vzduchu -> redkční podmínky -> pomalý rozklad organiky
- redukce Fe  a Mn -> šedý horizont s odstíny zelené a modré 
zóna trvalého zamokření vs kapilárního vzlínání
oxidační skvrny -- kanálky po kořenech
- Při kolísání hladiny/vzlínání se tvoří rezavé skvrny Fe (Goethit)

#### Pseudoglejový proces
Půda s vodou na povrchu nebo vázanou v horizontech se sníženou propustností -> redukční podmínky
výrazné mramorování

#### Podzolizační proces

humidní oblasti mírného a chladného pásma s produkcí kyselého lesního opadu (jehličnany) -> tvorba org. kyselin.
vybělený horizont
