## Cvičení 01 -- Vynášení strukturních prvků do Schmidtovy sítě

Cílem prvního cvičení je naučit se vynášení lineárních a plánárních prvky do Schmidtovy sítě. Schmidtova síť se také nazývá  Stereonet.

__Lineární prvek__ (*linear feature*) představuje strukturní prvek, který můžeme aproximovat přímkou.

__Planární prvek__ (*planar feature*) představuje strukturní prvek, který můžeme aproximovat plochou. 

Přehled lineárních a plánárních strukturních prvků:

| název č.  | název e.    | typ prvku |
|:----------|:------------|:----------|
| osa vrásy | *fold axis* | lineární  |
| stopa vrásového ramena |...| lineární|
| stopa osní roviny | ... | lineární |


### Zobrazení lineárního prvku

Při projekci ho zobrazujeme pomocí vektoru jehož obrazem je ve stereonetu bod.

```python
from apsg import Lin

sn = StereoNet()
lins = [
    Lin(0, 45),
    Lin(45, 45),
    Lin(90, 45),
    Lin(135, 45),
    Lin(180, 45),
    Lin(225, 45),
    Lin(270, 45), 
    Lin(315, 45),
]

for lin in lins:
    sn.line(lin, label=str(lin))
```

## Zobrazení planárního prvku 

Při projekci mlžeme zobrazit dvěma způsoby:

1. jako vektor představující pól plochy
2. jako vektor udávající směr spádu a spád plochy.

### 1. Zobrazení pomocí pólu plochy


```python
from apsg import Fol

sn = StereoNet()

fols = [
    Fol(0, 45),
    Fol(45, 45),
    Fol(90, 45),
    Fol(135, 45),
    Fol(180, 45),
    Fol(225, 45),
    Fol(270, 45), 
    Fol(315, 45),
]

for fol in fols:
    sn.pole(fol, label=str(fol))
```

### 2. Zobrazení pomocí průsečnice

```python
sn = StereoNet()

for fol in fols:
    sn.plane(fol, label=str(fol))
```

```python
sn = StereoNet()

"""
fols = [
    Fol(125, 35),
    Fol(270, 45),
    Fol(42, 27),
    Fol(180, 90),
]
"""
colors = "red", "green", "blue"

for i, fol in enumerate(fols[:3]):
    sn.plane(fol, color=colors[i])
    sn.pole(fol, color=colors[i])
```

## 3. Zdokumentuje a zobrazte vrásu

```python
s = StereoNet()

r1 = Fol(70, 71)
r2 = Fol(258, 55)

s.plane(r1)
s.plane(r2)

fold_axis = r1 ** r2

s.line(fold_axis)
```
