## Voda a $\delta ^{18}O$

Veličina $\delta ^{18}O$ a její pozitivní respektive negativní hodnoty určují odchylku od izotopického složení kyslíku ve standardu SMOW.

- $$\delta ^{18}O [promile] = \frac{R_{vzr}}{R_{std} - 1} 1000$$ 
- $$R = \frac{\delta^{18}O}{\delta ^{16}O}$$
- Pozitivní odchylka ukazuje na obohacení těžším izotopem $^{18}O$ (metamorfni a magmatická voda).
- Negativní odchylka ukazuje na ochuzení těžším izotopem $^{16}O$.

> Disociační konstanta = pro kyseliny nebo zásady lze kvantitativně určit "schopnost" odevzat nebo přijmout volné protony.
