```python
# pure shear
alpha = radians (45)
R = array ([[cos(alpha), -sin(alpha)], [sin(alpha), cos(alpha)]])
S = array ([[2, 0], [0,0.5]])
```

```python
F = R @ S @ R.T
def_show(F)
```

```python
u, s, v = svd(F)
```

```python
s # vrací jen diagonálu 
```

```python
diag(s) # udělá diagonální matici z s 
```

```python
u @ diag(s) @ v #to, co nám vyšlo 
```

```python
ar = []
gamas = linspace(0, 10, 50)
for gama in gamas: 
    F = array([[1, gama],[0,1]])
    u, s, v = svd(F)
    ar.append(s[0]/s[1])
plot (gamas, ar)
```

```python
ori = []
gamas = linspace(0, 10, 50)
for gama in gamas: 
    F = array([[1, gama],[0,1]])
    u, s, v = svd(F)
    ori.append(degrees(arctan2(u[1,0], u[0,0])))
plot (gamas, ori)
```
