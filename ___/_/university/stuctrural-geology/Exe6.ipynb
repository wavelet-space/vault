```python
%pylab inline
from scipy.linalg import expm,logm
```

```python
F = ([[2,2],[0,0.5]])
L = logm(F)
times= linspace (0.001,4,100)
```

```python
ar = []
alphas = []
```

```python
for t in times: 
    Fi = expm (L*t)
    U,s,V = svd(Fi)
    ar.append(s[0]/s[1])
    alphas.append(degrees(arctan2(U[1,0], U[0,0])))
```

```python
plot(times,ar) ;
```

Exe_2: viz presentace slide 70 - transprese

```python
edot = 1e-14 #puresheare hodnota
gdot = 1e-14
```

```python
L = array ([[0,gdot,0],
            [0,-edot,0],
            [0,0,edot]])            
```

```python
k,d = [],[]
Ryz=[]
Rxy=[]
times=linspace (0,5) 
yearsec = 365.25*24*3600
```

```python
for t in times:
    F=expm(L*t*1e6*yearsec)
    U,s,V = svd(F)
    rxy = s[0]/s [1]
    ryz = s[1]/s[2]
    d.append ( sqrt((rxy-1)**2 + (ryz-1)**2))
    k.append ((rxy -1) / (ryz-1))
    Rxy.append (rxy)
    Ryz.append(ryz)
```

```python
plot (times, d)
```

```python
plot (times, k) #symetrie , viz flyn diagram k<1
```

```python
plot (Ryz,Rxy)
axis('equal') ;
```
