```python
theta = linspace(0, 2*pi, 300)
Xc, Yc = cos(theta), sin(theta)
plot(Xc, Yc, 'g') 

F = array([[1,1], [0, 1]])
xe, ye = dot(F, [Xc, Yc])  #pokud sem nedám dva členy před rovnítko, vyleze mi matice ---> takto se matice rozdělí na její dvě komponenty 
plot(xe, ye, 'r')
axis('equal');
```

```python
# coordinates of square
Xs = [1, -1, -1, 1, 1]   # definován na tvrdo
Ys = [1, 1, -1, -1, 1]
plot(Xs, Ys, 'g')

# Apply deformation gradient and plot result
F = array([[2, 0], [0, 0.5]])   # matice pro deformaci 
xn, yn = dot(F, [Xs, Ys])
plot(xn, yn, 'r')
axis('equal');
```

```python
# create rectangular grid
X, Y = meshgrid(linspace(-2.2, 2.2, 21), linspace(-1.9, 1.9, 17)) 

# Apply deformation gradient. As [X, Y] is 3D array, we need
# special function to treat dot product properly. tensordot
# can do it.
x, y = tensordot(F, [X, Y], axes=1)  # násobení matice s vícerozměrnou maticí ---> technická zkratka

# plot displacement vectors on all point
quiver(X, Y, x-X, y-Y, angles='xy') # fce pro vykreslení vektorových polí, kdy X,Y jsou pův. souřadnice a x-X je vektor přemístění
plot(Xc, Yc, 'g', xe, ye, 'r')
plot(Xs, Ys, 'g', xn, yn, 'r')
axis('equal');
```

```python
# Tady rovnou vypočtu komponenty a pak to dosadím :

# calculate displacements
J = F - eye(2)
u, v = tensordot(J, [X, Y], axes=1)

# plot
quiver(X, Y, u, v, angles='xy')
plot(Xc, Yc, 'g', xe, ye, 'r')
plot(Xs, Ys, 'g', xn, yn, 'r')
axis('equal');
```
