Př.1: Pro jaký úhel je jaké protažení? - změna velikosti vektoru 

```python
F = array ([[2,1],[0,0.5]])  # definujeme deformacni gradient
F
```

```python
C = F.T @ F   # vypočteme greenuv def tenzor
C
```

```python
theta = linspace(0,2*pi,100)
S = []
for t in theta: 
    u = array ([cos(t), sin(t)])
    S.append(sqrt(u @ C @ u)) 
```

```python
plot(degrees(theta), S)
axhline(1,color='r');
```

Př. 3: graf pro změnu pozičního vektoru 

```python
F = array ([[1,0],[1,1]])
C = F.T @ F
theta = linspace(0,pi,100)
phi = []  #uhel po def
```

```python
for t in theta:
    X1 = array ([cos(t-pi/4), sin(t-pi/4)])
    X2 = array ([cos(t+pi/4), sin(t+pi/4)])
    phi.append(arccos( (X1 @ C @ X2)/(sqrt(X1 @ C @ X1) * sqrt(X2 @ C @ X2)) ))
```

```python
plot(degrees(theta), degrees(phi))
axhline(90,color='red');
```
